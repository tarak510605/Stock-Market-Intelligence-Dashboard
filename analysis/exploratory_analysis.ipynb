`````python
````xml
<VSCode.Cell language="markdown">
# Exploratory Data Analysis
## Stock Market Intelligence Dashboard

**Project**: Stock Market Intelligence Dashboard  
**Purpose**: Analyze stock price trends, trading volume, and stock performance  
**Date**: March 2026

This notebook performs exploratory data analysis on historical stock data for:
- Apple (AAPL)
- Microsoft (MSFT)
- Tesla (TSLA)
- Nvidia (NVDA)
</VSCode.Cell>

<VSCode.Cell language="code">
# Import required libraries
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import sys
import warnings
warnings.filterwarnings('ignore')

# Add parent directory to path for imports
sys.path.append('../')

# Import custom modules
from utils.helper_functions import load_data, create_summary_table
from visualization.charts import (plot_stock_prices, plot_volume_analysis, 
                                  plot_daily_returns, plot_correlation_heatmap,
                                  plot_cumulative_returns)

# Display settings
pd.set_option('display.max_columns', None)
pd.set_option('display.width', 1000)
%matplotlib inline

print("✓ Libraries imported successfully")
</VSCode.Cell>

<VSCode.Cell language="markdown">
## 1. Load Data

Load the cleaned stock data that was prepared by the ETL pipeline.
</VSCode.Cell>

<VSCode.Cell language="code">
# Load cleaned data
data = load_data('../data/cleaned/stock_data_cleaned.csv')

if data is not None:
    print(f"✓ Data loaded successfully")
    print(f"Shape: {data.shape}")
    print(f"\nColumns: {', '.join(data.columns.tolist())}")
    print(f"\nSymbols: {', '.join(data['symbol'].unique())}")
    print(f"Date range: {data['date'].min()} to {data['date'].max()}")
else:
    print("✗ Failed to load data. Please run ETL scripts first.")
</VSCode.Cell>

<VSCode.Cell language="markdown">
## 2. Data Overview

Examine the first few rows and basic statistics of the dataset.
</VSCode.Cell>

<VSCode.Cell language="code">
# Display first few rows
print("="*80)
print("FIRST 10 ROWS")
print("="*80)
display(data.head(10))
</VSCode.Cell>

<VSCode.Cell language="code">
# Display data info
print("\n" + "="*80)
print("DATA INFO")
print("="*80)
data.info()
</VSCode.Cell>

<VSCode.Cell language="code">
# Display summary statistics
print("\n" + "="*80)
print("SUMMARY STATISTICS")
print("="*80)
display(data.describe())
</VSCode.Cell>

<VSCode.Cell language="markdown">
## 3. Summary Table by Stock

Create a comprehensive summary table for each stock.
</VSCode.Cell>

<VSCode.Cell language="code">
# Create and display summary table
summary_table = create_summary_table(data)

print("="*80)
print("STOCK SUMMARY TABLE")
print("="*80)
display(summary_table)
</VSCode.Cell>

<VSCode.Cell language="markdown">
## 4. Stock Price Trends Analysis

Visualize how stock prices have changed over time.
</VSCode.Cell>

<VSCode.Cell language="code">
# Plot stock price trends for all stocks
plot_stock_prices(data, save_path='../data/processed/price_trends.png')
</VSCode.Cell>

<VSCode.Cell language="markdown">
### Observations:

Analyze the price trends chart:
- Which stock showed the highest growth?
- Which stock is most volatile?
- Are there any notable patterns or events visible in the data?
</VSCode.Cell>

<VSCode.Cell language="markdown">
## 5. Trading Volume Analysis

Analyze trading volume patterns for each stock.
</VSCode.Cell>

<VSCode.Cell language="code">
# Analyze volume for AAPL
print("Analyzing Apple (AAPL) Trading Volume")
plot_volume_analysis(data, 'AAPL', save_path='../data/processed/volume_aapl.png')
</VSCode.Cell>

<VSCode.Cell language="code">
# Analyze volume for MSFT
print("Analyzing Microsoft (MSFT) Trading Volume")
plot_volume_analysis(data, 'MSFT', save_path='../data/processed/volume_msft.png')
</VSCode.Cell>

<VSCode.Cell language="code">
# Analyze volume for TSLA
print("Analyzing Tesla (TSLA) Trading Volume")
plot_volume_analysis(data, 'TSLA', save_path='../data/processed/volume_tsla.png')
</VSCode.Cell>

<VSCode.Cell language="code">
# Analyze volume for NVDA
print("Analyzing Nvidia (NVDA) Trading Volume")
plot_volume_analysis(data, 'NVDA', save_path='../data/processed/volume_nvda.png')
</VSCode.Cell>

<VSCode.Cell language="markdown">
## 6. Daily Returns Analysis

Examine the distribution of daily returns to understand volatility patterns.
</VSCode.Cell>

<VSCode.Cell language="code">
# Plot daily returns distribution
plot_daily_returns(data, save_path='../data/processed/daily_returns_dist.png')
</VSCode.Cell>

<VSCode.Cell language="code">
# Calculate and display return statistics
print("="*80)
print("DAILY RETURN STATISTICS BY STOCK")
print("="*80)

for symbol in data['symbol'].unique():
    symbol_data = data[data['symbol'] == symbol]
    returns = symbol_data['daily_return'].dropna()
    
    print(f"\n{symbol}:")
    print(f"  Mean Return: {returns.mean():.4f}%")
    print(f"  Median Return: {returns.median():.4f}%")
    print(f"  Std Deviation: {returns.std():.4f}%")
    print(f"  Min Return: {returns.min():.4f}%")
    print(f"  Max Return: {returns.max():.4f}%")
    print(f"  Positive Days: {(returns > 0).sum()} ({(returns > 0).sum() / len(returns) * 100:.1f}%)")
    print(f"  Negative Days: {(returns < 0).sum()} ({(returns < 0).sum() / len(returns) * 100:.1f}%)")
</VSCode.Cell>

<VSCode.Cell language="markdown">
## 7. Cumulative Returns Comparison

Compare how $100 invested in each stock would have grown over time.
</VSCode.Cell>

<VSCode.Cell language="code">
# Plot cumulative returns
plot_cumulative_returns(data, save_path='../data/processed/cumulative_returns.png')
</VSCode.Cell>

<VSCode.Cell language="code">
# Calculate final cumulative returns
print("="*80)
print("CUMULATIVE RETURNS (Total % Gain/Loss)")
print("="*80)

for symbol in data['symbol'].unique():
    symbol_data = data[data['symbol'] == symbol]
    final_return = symbol_data['cumulative_return'].iloc[-1] * 100
    print(f"{symbol}: {final_return:+.2f}%")
</VSCode.Cell>

<VSCode.Cell language="markdown">
## 8. Stock Correlation Analysis

Analyze how stocks move in relation to each other.
</VSCode.Cell>

<VSCode.Cell language="code">
# Plot correlation heatmap
correlation_matrix = plot_correlation_heatmap(data, save_path='../data/processed/correlation_heatmap.png')

print("\n" + "="*80)
print("CORRELATION MATRIX")
print("="*80)
display(correlation_matrix)
</VSCode.Cell>

<VSCode.Cell language="markdown">
### Correlation Insights:

- **High Correlation (> 0.8)**: Stocks that tend to move together
- **Low Correlation (< 0.3)**: Stocks that move independently
- **Negative Correlation (< 0)**: Stocks that move in opposite directions

This information is valuable for portfolio diversification.
</VSCode.Cell>

<VSCode.Cell language="markdown">
## 9. Price Range Analysis

Analyze the daily price range (High - Low) to understand intraday volatility.
</VSCode.Cell>

<VSCode.Cell language="code">
# Calculate average price range by stock
print("="*80)
print("AVERAGE DAILY PRICE RANGE")
print("="*80)

for symbol in data['symbol'].unique():
    symbol_data = data[data['symbol'] == symbol]
    avg_range = symbol_data['price_range'].mean()
    avg_range_pct = (avg_range / symbol_data['close'].mean()) * 100
    
    print(f"{symbol}:")
    print(f"  Average Range: ${avg_range:.2f}")
    print(f"  As % of Price: {avg_range_pct:.2f}%")
</VSCode.Cell>

<VSCode.Cell language="markdown">
## 10. Year-over-Year Analysis

Compare stock performance across different years.
</VSCode.Cell>

<VSCode.Cell language="code">
# Add year column if not exists
if 'year' not in data.columns:
    data['year'] = pd.to_datetime(data['date']).dt.year

# Calculate yearly returns by stock
yearly_performance = []

for symbol in data['symbol'].unique():
    symbol_data = data[data['symbol'] == symbol]
    
    for year in symbol_data['year'].unique():
        year_data = symbol_data[symbol_data['year'] == year]
        
        if len(year_data) > 1:
            year_return = ((year_data['close'].iloc[-1] / year_data['close'].iloc[0]) - 1) * 100
            
            yearly_performance.append({
                'Symbol': symbol,
                'Year': year,
                'Return (%)': year_return,
                'Start Price': year_data['close'].iloc[0],
                'End Price': year_data['close'].iloc[-1]
            })

yearly_df = pd.DataFrame(yearly_performance)

print("="*80)
print("YEARLY PERFORMANCE SUMMARY")
print("="*80)
display(yearly_df.pivot(index='Year', columns='Symbol', values='Return (%)'))
</VSCode.Cell>

<VSCode.Cell language="markdown">
## 11. Key Insights Summary

Based on the exploratory data analysis, we can draw several conclusions:

### Price Trends
- Identify which stock had the highest overall return
- Note any significant price movements or patterns

### Volatility
- Compare daily return standard deviations
- Identify most and least volatile stocks

### Correlation
- Stocks with high correlation move together (less diversification benefit)
- Stocks with low correlation provide better diversification

### Volume Patterns
- High volume days often coincide with significant price movements
- Volume trends can indicate investor interest

### Performance
- Cumulative returns show total investment growth
- Year-over-year analysis reveals consistency

**Next Steps**: 
- Technical indicator analysis (moving averages, RSI, etc.)
- Portfolio optimization
- Risk assessment
</VSCode.Cell>

<VSCode.Cell language="markdown">
## 12. Export Analysis Results

Save key findings for reporting and dashboard creation.
</VSCode.Cell>

<VSCode.Cell language="code">
# Export summary table to CSV
summary_table.to_csv('../data/processed/stock_summary.csv', index=False)
print("✓ Summary table exported to: ../data/processed/stock_summary.csv")

# Export yearly performance
yearly_df.to_csv('../data/processed/yearly_performance.csv', index=False)
print("✓ Yearly performance exported to: ../data/processed/yearly_performance.csv")

# Export correlation matrix
correlation_matrix.to_csv('../data/processed/correlation_matrix.csv')
print("✓ Correlation matrix exported to: ../data/processed/correlation_matrix.csv")
</VSCode.Cell>

<VSCode.Cell language="markdown">
## Conclusion

This exploratory analysis provided a comprehensive overview of the stock data including:
- Price trends and patterns
- Trading volume behavior  
- Return distributions and volatility
- Stock correlations
- Year-over-year performance

These insights form the foundation for more advanced technical analysis and portfolio optimization.

---
**End of Exploratory Data Analysis**
</VSCode.Cell>
````
`````
